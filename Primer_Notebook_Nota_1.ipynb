# ✅ Problema de Ingeniería (Nivel Básico)

### 📌 Cálculo del volumen de concreto para una losa rectangular

En ingeniería civil es común calcular cuánto concreto se necesita para una losa.

La fórmula es:

[
Volumen = Largo \times Ancho \times Espesor
]

---

# 1️⃣ Definición del algoritmo (Entradas – Proceso – Salidas)

## 🔹 Entradas

* Largo de la losa (m)
* Ancho de la losa (m)
* Espesor de la losa (m)

## 🔹 Proceso

* Multiplicar largo × ancho × espesor

## 🔹 Salida

* Volumen de concreto en metros cúbicos (m³)

---

# 2️⃣ Diseño del algoritmo paso a paso

### 📌 Lógica de solución

1. Pedir al usuario el largo.
2. Pedir el ancho.
3. Pedir el espesor.
4. Calcular el volumen.
5. Mostrar el resultado.

---

# 3️⃣ Pseudocódigo

```
Inicio
    Leer largo
    Leer ancho
    Leer espesor
    volumen ← largo * ancho * espesor
    Mostrar volumen
Fin
```

Observa que aquí ya estamos usando:

* Variables
* Operador de multiplicación (*)
* Asignación (←)

---

# 4️⃣ Implementación en Python

```python
# -----------------------------------------
# Cálculo del volumen de una losa de concreto
# Programa básico para estudiantes de Ingeniería
# -----------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo de la losa en metros: "))
ancho = float(input("Ingrese el ancho de la losa en metros: "))
espesor = float(input("Ingrese el espesor de la losa en metros: "))

# Proceso
volumen = largo * ancho * espesor

# Salida
print("\nEl volumen de concreto necesario es:", volumen, "m³")
```

---

# 5️⃣ Explicación de variables y operadores usados

### 🔹 Variables

* `largo`
* `ancho`
* `espesor`
* `volumen`

Son espacios en memoria donde almacenamos datos.

### 🔹 Operadores

* `=` → Asignación
* `*` → Multiplicación
* `float()` → Conversión a número decimal

---

# 6️⃣ Ejemplo de ejecución

Si el usuario ingresa:

```
Largo = 10
Ancho = 5
Espesor = 0.15
```

El programa calcula:

[
10 \times 5 \times 0.15 = 7.5 , m^3
]

### Un ejemplo de lógica aplicada

Antes de escribir código, recuerda que en ingeniería siempre partimos de un fenómeno o necesidad. Imagina que queremos calcular el **volumen de agua necesario para un riego** basado en el área de un cultivo.

Para mostrarte cómo se ve Python, aquí tienes un script sencillo. Nota que he usado **nombres de variables descriptivos**, algo vital para que tus colegas entiendan tus programas:

````python
# Ejemplo: Cálculo de volumen de riego
def calcular_volumen_riego(area_m2, lamina_riego_mm):
    """
    Calcula el volumen de agua necesario en metros cúbicos.
    1 mm de lámina equivale a 1 litro por metro cuadrado.
    """
    # Convertimos lámina (mm) a metros para el cálculo
    volumen_m3 = area_m2 * (lamina_riego_mm / 1000)
    return volumen_m3

# Datos de entrada
area_cultivo = 5000  # metros cuadrados
necesidad_hidrica = 5 # mm de lámina de agua

# Llamada a la función
total_agua = calcular_volumen_riego(area_cultivo, necesidad_hidrica)

print(f"Para un área de {area_cultivo} m2, se requieren {total_agua} m3 de agua.")
````
