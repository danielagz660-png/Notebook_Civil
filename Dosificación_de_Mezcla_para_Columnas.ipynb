# **Dosificación de Mezcla para Columnas**
Planteamiento:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular. Se sabe que para un concreto estándar de 210 kg/cm2
(3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$v = \pi * r^2 * h$$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$sacos = v * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3

Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).

```python
# ENTRADA:1. Solicitar la altura de la columna en metros.
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el adio r, calculael volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: Imprime el resultado
print(f"El volumen de la columna es: {v:.2f} m3")
print(f"Se requieren {sacos:.2f} sacos de cemento")
```

Este código tiene como objetivo calcular la cantidad de sacos de cemento necesarios para una columna de sección circular, basándose en su altura y diámetro.

### Explicación Línea por Línea del Código de Cálculo de Sacos de Cemento:

1.  `h = float(input("Ingrese la altura de la columna en metros: "))`
    *   **Propósito:** Solicita al usuario la altura de la columna.
    *   **Funcionamiento:**
        *   `input("...")`: Muestra el mensaje entre comillas al usuario y espera que ingrese un valor por teclado.
        *   `float(...)`: Convierte el texto ingresado por el usuario a un número decimal (punto flotante), ya que las dimensiones pueden ser no enteras.
        *   `h = ...`: Asigna el valor numérico ingresado a la variable `h` (altura).

2.  `d = float(input("Ingrese el diámetro de la columna en metros: "))`
    *   **Propósito:** Solicita al usuario el diámetro de la columna.
    *   **Funcionamiento:** Similar a la línea anterior, pero para el diámetro `d`.

3.  `r = d/2`
    *   **Propósito:** Calcula el radio de la columna.
    *   **Funcionamiento:** Divide el diámetro `d` entre 2 y asigna el resultado a la variable `r` (radio).

4.  `v = 3.1416 * r**2 * h`
    *   **Propósito:** Calcula el volumen de la columna cilíndrica.
    *   **Funcionamiento:** Aplica la fórmula del volumen de un cilindro (π * radio² * altura).
        *   `3.1416`: Es una aproximación del valor de Pi (π).
        *   `r**2`: Calcula el radio `r` elevado al cuadrado.
        *   `*`: Son operadores de multiplicación.
        *   `v = ...`: Asigna el resultado del cálculo a la variable `v` (volumen).

5.  `sacos = v * 9.5`
    *   **Propósito:** Calcula la cantidad total de sacos de cemento necesarios.
    *   **Funcionamiento:** Multiplica el volumen `v` de la columna (en metros cúbicos) por la cantidad de sacos de cemento requeridos por cada metro cúbico (9.5 sacos/m³).
        *   `sacos = ...`: Asigna el resultado a la variable `sacos`.

6.  `print(f"El volumen de la columna es: {v:.2f} m3")`
    *   **Propósito:** Muestra en pantalla el volumen calculado de la columna.
    *   **Funcionamiento:**
        *   `f"..."`: Es una f-string (cadena formateada) que permite incrustar variables y expresiones directamente dentro de la cadena.
        *   `{v:.2f}`: Inserta el valor de la variable `v` y lo formatea para que tenga dos decimales.
        *   `print(...)`: Imprime la cadena resultante en la consola.

7.  `print(f"Se requieren {sacos:.2f} sacos de cemento")`
    *   **Propósito:** Muestra en pantalla la cantidad de sacos de cemento requeridos.
    *   **Funcionamiento:** Similar a la línea anterior, imprime el valor de `sacos` formateado a dos decimales.
